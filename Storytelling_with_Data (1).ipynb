```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
sales_df = pd.DataFrame({"Month": ['Sep-22', 'Oct-22', 'Nov-22', 'Dec-22', 'Jan-23', 'Feb-23', 'Mar-19'],
                        "Sales": [10890, 11049, 11190, 11209, 11489, 10900, 11490]})
sales_df.head()
```

```python
fig = px.line(sales_df, x = "Month", y = "Sales",
              title='Huge Dip in Sales After Period of Continued Growth', color_discrete_sequence=["#808080"])
fig.update_layout(autosize=False, width=700, height=400)
fig.show()
```

```python
fig = px.line(sales_df, x = "Month", y = "Sales", text="Sales",
              title='<span style="color: red;"><b>Huge Dip</b></span> in Sales <br>After Period of Continued Growth', color_discrete_sequence=["#808080"])
fig.update_traces(textposition="bottom right")
# Specify the portion of the line chart you want to change the color for
start_index = 4
end_index = 6

# Create a new line trace for the specified portion and set the color
x_segment = sales_df['Month'].values[start_index:end_index + 1]
y_segment = sales_df['Sales'].values[start_index:end_index + 1]
fig.add_trace(go.Scatter(x=x_segment, y=y_segment, mode='lines', name='Huge Dip in Sales', line=dict(color='red')))

fig.add_vline(x='Sep-22', line_width=1, line_dash="dash", line_color="gray")
fig.add_annotation(
    text="Period of growth",
    x='Oct-22',
    y=11489,
    showarrow=False,  # Set to False if you don't want an arrow pointing to the text
)
fig.update_layout(plot_bgcolor='rgba(0,0,0,0)')
fig.update_yaxes(visible=False, showticklabels=False)
fig.update_layout(autosize=False, width=700, height=400)
fig.show()
```
